```python
from bs4 import BeautifulSoup
```

```python
import pandas as hs
```

## Scrapping Websie

```python
url = "https://www.flipkart.com/search?q=computer&sid=6bo%2Cnl4%2Cigk&as=on&as-show=on&otracker=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_HistoryAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=computer%7CAll+In+One+PCs&requestId=9bb2cb93-3a40-4bbb-8f27-ad8f7c621965&as-searchtext=co"
```

## Website Response

```python
res = requests.get(url)
```

```python
res
```

## parsing

```python
soup = BeautifulSoup(res.text)
```

```python
soup
```

## computer names

```python
computer=soup.select(".s1Q9rs")
```

```python
computer
```

```python
prices = soup.select('div~ div+ div ._4ddWXP ._30jeq3 , .col-12-12:nth-child(3) ._30jeq3 , ._13oc-S div:nth-child(1) ._4ddWXP ._30jeq3')
```

```python
prices
```

```python
price_title=[]
```

```python
for item in prices:
    price_title.append(item.text)
```

```python
price_title
```

```python
computer_names=[]
```

```python
for item in computer:
    computer_names.append(item.text)
```

```python
computer_names
```

## Data Frames

```python
df=hs.DataFrame(computer_names,columns=['computer names'])
```

```python
df
```

```python
df['price_title']=price_title
```

```python
df
```

```python
df.to_csv('hr.csv')
```
